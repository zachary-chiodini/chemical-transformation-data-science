```python
import import_ipynb, pandas as pd
from time import sleep
from crossref import Crossref
```

```python
measures = pd.read_csv( 'measures.tsv', sep = '\t', dtype = str )
measures = measures.fillna( '' )
measures[ 'doi' ] = ''
measures = measures[ [ col for col in measures.columns.tolist() if  col != 'Unnamed: 0' and col != 'id' ] ]
```

```python
crossref = Crossref()
```

```python
citations = pd.DataFrame()
for indx, cite in measures[ 'citation' ].items() :
    get = pd.DataFrame( crossref.citation( cite ) )
    get[ 'index' ] = indx
    citations = pd.concat( [ citations, get ], axis = 0 )
    measures.loc[ indx, 'doi' ] = get[ 'doi' ].iloc[ 0 ]
citations = citations.reset_index( drop = True )
citations = citations.fillna( '' )
```

```python
citations = citations[ ~ citations[ [ col for col in citations.columns if col != 'index' ] ].duplicated( keep = 'first' ) ]
```
